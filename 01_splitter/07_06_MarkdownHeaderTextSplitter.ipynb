## 1.MarkdownHeaderTextSplitter

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter

```

# Splitter 생성

```python
headers_to_split_on = [
    (
        "#",
        "Header 1"
        ),
    (
        "##",
        "Header 2"
        ),
    (
        "###",
        "Header 3"
        ),
]
```

```python
markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on = headers_to_split_on)
```

```python
# Assuming your markdown text is:
markdown_text = """
# Header 1
This is some text under Header 1.

## Header 2
This is some text under Header 2.

### Header 3
This is some text under Header 3.
"""

# Now you can call split_text() with the markdown_text:
md_header_splits = markdown_splitter.split_text(text=markdown_text)
```

```python
for header in md_header_splits:
  print(f'{header.page_content}')
  print(f'{header.metadata}', end='\n=============')
```

```python
markdown_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on = headers_to_split_on,
    strip_headers=False,
)

md_header_splits = markdown_splitter.split_text(text=markdown_text)

for header in md_header_splits:
  print(f'{header.page_content}')
  print(f'{header.metadata}', end='\n=============')
```
